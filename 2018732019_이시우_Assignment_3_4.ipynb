### 4 실습 (Assignment 3)
* 행렬 방정식 풀기  
  * 다음 행렬 방정식을 'Pseudo inverse matrix'를 이용해 풀어보자  
  * A^T A의 역행렬이 존재한다고 가정 


```python
import numpy as np
list_A = np.array([[0, 1], [1, 1], [2, 1], [3, 1]])
list_B = np.array([-1, 0.2, 0.9, 2.1])
```

```python
import numpy.linalg as lin
```

```python
list_inv_A = lin.pinv(list_A)
print(list_inv_A)
```

```python
list_inv_AB = np.matmul(list_inv_A, list_B)
```

```python
print(list_inv_AB)
```
